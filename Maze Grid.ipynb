```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
```

```python
x_axis, y_axis, colors = [], [], []

for x in range(0, 10):
    for y in range(0, 10):
        up = np.random.randint(0, 2)
        x_axis.append([x, x + .8])
        y_axis.append([y + .8 * (up >= .5), y + .8 * (up < .5)])
        col = np.random.randint(0, 4)
        colors.append({0: 'blue', 1: 'red', 2: 'green', 3: 'yellow'}[col])
```

```python
output_notebook()
```

```python
plot = figure(plot_width=800, plot_height=800)

plot.grid.visible = False
plot.xaxis.visible = False
plot.yaxis.visible = False

plot.multi_line(x_axis, y_axis, color=colors, line_width=6, line_alpha=.1, line_cap='round')
plot.multi_line(x_axis, y_axis, color='black')

show(plot)
```
